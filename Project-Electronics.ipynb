# Load the dataset into pandas and provide a summary of columns in the loaded dataset without any modification.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
df=pd.read_csv(r"C:\Users\acer\Desktop\electronics.csv")
df
```

```python
df.info()
```

# Prepare the data 

# Step 1 check for null values, drop it and cross check 


```python
df1=df.dropna()
```

```python
df1.isnull().sum()
```

# Step 2 check for duplicate values and drop it


```python
df2=df1.drop_duplicates()
df2
```

# Step 3 rectify column name

```python
df2.columns=['Order_ID','Product','Quantity_Ordered','Price_Each','Order_Date','Purchase_address']
df2.head
```

# Step 4 Extract date from order date

```python
df2['Date']=df2['Order_Date'].str[:8]
df2
```

```python
df2.info()
```

# Step 5 extract month from order date


```python
df2['Month']=df2['Date'].str[:2]
df2
```

# Step 6 extract house_no ,city and pin code from address


```python
df2[['House_No','City','Pin_Code']]=df2.Purchase_address.str.split(",",expand=True)
df2
```

# Step 7 Check for month column for bad data entry and remove it

```python
df2.Month.value_counts()
```

```python
df2=df2[df2['Month'].str[:2] !='Or']
df2.head()
```

```python
df2.shape
```

# Step 8 correct data types

```python
df2['Month']=df2['Month'].astype('int32')
df2['Order_Date']=pd.to_datetime(df2['Order_Date'])
df2['Date']=pd.to_datetime(df2['Date'])
df2.info()
```

# Step 9 make Sales Column (hint Quantity_ordered * Price_each)

```python
df2['Quantity_Ordered']=df2['Quantity_Ordered'].astype('int32')
df2['Price_Each']=df2['Price_Each'].astype('float64')
df2.info()
```

```python
df2['Sales']=df2['Quantity_Ordered']*df2['Price_Each']
df2
```

#  Sales Value by Each Month

Monthly_sales= df2.groupby('Month').Sales.sum()
Monthly_sales

# City had the highest sales overall

```python
city_sales=df2.City.value_counts()
city_sales
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=df2['city'])
plt.xlabel("City name")
plt.ylabel("Number of sells")
plt.title("City vs sales")
```

# Distribution of sales over different times of the day (Morning, Afternoon, Evening, Night)

```python
df2['Hour']=df2.Order_Date.dt.hour
```

```python
df2.groupby('Hour').count()
```

```python
hours=[hour for hour , df2 in df2.groupby('Hour')]
plt.plot(hours, df2.groupby('Hour').count())
plt.grid()
plt.xticks(hours)
plt.xlabel('Hours in 24 hr format')
plt.ylabel('Orders')
plt.show()
```

# Highest Product Sold by Amount and Numbers

```python
highest_orders=df2.Product.value_counts()
highest_sales=df2.groupby('Product').Sales.sum().sort_values(ascending=False)
fig,axes=plt.subplots(1,2,figsize=(20,7))
sns.barplot(ax=axes[0],x=highest_orders.index,y=highest_orders.values)
axes[0].set_xlabel('Products')
axes[0].set_ylabel('No of Orders')
axes[0].tick_params(axis='x',rotation=90)
sns.barplot(ax=axes[1],x=highest_sales.index,y=highest_sales.values)
axes[1].set_xlabel('Products')
axes[1].set_ylabel('Sales Amount')
axes[1].tick_params(axis='x',rotation=90)
```
